# Ações e Séries Temporais

Aqui será feito um estudo de séries temporais usando dados de ações. Começamos importando bibliotecas relevantes e o conjunto de dados, que será um conjunto dos valores das ações do Banco do Brasil de 2016 para cá. Note que não há dados faltantes.

```python
#----------------------------# bibliotecas: 
#--------------------------------------# 
import numpy           as   np         #
import pandas           as   pd        #
import vectorbt          as   vbt      #
import seaborn            as   sns     #
import yfinance            as   yf     #
import stats_tests          as   st    #
import financefunction       as   ff   #
import matplotlib.pyplot      as   plt #
#--------------------------------------#

#-------------------# prepara visualização:
%matplotlib inline
sns.set() 
#---------------------------------------# dataset: 
#------------------------------------------------------------------#
t0 = pd.to_datetime('2016-01-01')                                  #
t1 = pd.to_datetime('now')                                         #
papel = yf.download('BBAS3.SA', start=t0, end=t1, progress=False)  #
#------------------------------------------------------------------#

papel.isnull().sum()
```

### Gráfico 1: 
Abaixo temos o gráfico dos valores de fechamento das ações de 2016 até hoje. 

```python
ff.dataplot([papel['Close']], ['fechamento'], ylabel='valor (R$)')
```

Não podemos decidir se uma série temporal é estacionária ou não apenas olhando para seu gráfico. Vamos fazer aqui dois testes estatísticos.

# Teste de Dickey-Fuller Aumentado (ADF):
Este teste estatístico ajuda a determinar se uma série temporal é estacionária ou não. Ele trabalha com as seguintes hipóteses: 
 - Hipótese Nula:        Série possui Raíz Unitária
 - Hipótese Alternativa: Série não possui Raíz Unitária
 
 Ele rejeita a Hipótese Nula no caso em que a estatística medida pelo teste é menor que o Valor Crítico e o p-valor é menor que 0.05.

```python
# A função abaixo está definido em stats_tests.py,
# e usa a biblioteca statsmodels.
st.adf_test(papel['Close'])
```

# Teste de Kwiatkowski–Phillips–Schmidt–Shin (KPSS):

Agora, vamos fazer o teste de KPSS. O fim do teste é o mesmo: ajudar a determinar se a série temporal é ou não estacionária. Mas aqui, ele trabalha com as seguintes hipóteses: 
- Hipótese Nula: A série é estacionária
- Hipótese Alternativa: A série possui uma raíz unitária. 

Ao contrário do teste visto acima, aqui, se o valor estatístico medido pelo teste for maior que o valor crítico, o p-Valor deverá será ser pequeno (digamos, menor que 0.05) e a hipótese nula é rejeitada, ou seja, é um indicativo de que a série é estacionária com tendência. 

```python
st.kpss_test(papel['Close'])
```

# Médias Móveis:

```python
ano2023 = (papel.index >= '2023') #---------> filtro


#--------------------# médias móveis:
#---------------------------------------------------------------------#
mms = [7, 50, 100]
for i in mms:                                               #
    papel['mm{}'.format(i)] = papel['Close'].rolling(window=i).mean() 
#---------------------------------------------------------------------#    


fig1 = plt.figure(figsize=(8,3), dpi=100, facecolor='moccasin')

axes1 = fig1.add_axes([0, 0, 1, 1])
axes1.plot(papel[ano2023]['Close'], 'k', label='fechamento')   
for i in mms:
    axes1.plot(papel[ano2023]['mm{}'.format(i)], label='mm{}'.format(i))   
axes1.set_ylabel('Valor (R$)')
axes1.legend(loc=0)
plt.show()
```
