# Comparação entre Renda Fixa e Variável:
Neste notebook, pretendo simular como teria sido investir 100 reais por mês em uma aplicação CDB de liquidez diária rendendo 100% do CDI, simular como teria sido a evolução de um patrimonio no mesmo período com compras mensais de algumas ações, considerar os dividendos, considerar a inflação, e comparar os resultados.



Feito: 
- Comparação gráfica entre valores parados em conta-corrente e valores investidos em renda-fixa

A fazer: 
- Evolução de um patrimônio pela estratégia de comprar e segurar ações;
- Estabelecer as métricas que vão medir o dinheiro investido;
- Descontar a inflação;
- Comparar as estratégias.

```python
#--------------------------------# bibliotecas
import requests                  #
import numpy         as np       #
import pandas         as pd      #
import seaborn         as sns    #
import yfinance         as yf    #
import matplotlib.pyplot as plt  #
#--------------------------------#

#-------------------# prepara visualização:
%matplotlib inline  
sns.set()           #
#-------------------#

#------------------------------------------------# ações de interesse:
ativos = [                                       #
    'NUBR33.SA', #   0  ->  nubank               #
    'BBAS3.SA',  #   1  ->  banco do brasil      #
    'AAPL34.SA', #   2  ->  apple                #
    'A1MD34.SA', #   3  ->  amd                  #
    'ITLC34.SA', #   4  ->  intel                #
    'GOGL34.SA'  #   5  ->  google               #
]                                                #
#------------------------------------------------#

#------------------------------------------------# 
t0 = pd.to_datetime('2016-01-01')                #
t1 = pd.to_datetime('now')                       #
papel = yf.download(ativos[1], start=t0, end=t1, actions=True) # --> dataset
#------------------------------------------------#

   
```

Abaixo, temos uma função que consulta uma API do Banco Central e retorna dados solicitados:

```python
hoje = pd.to_datetime("today").strftime("%d/%m/%Y")

def get_bc(codigo, data_inicial=None, data_final=hoje):
    url = 'https://api.bcb.gov.br/dados/serie/bcdata.sgs.{}/dados?formato=json'.format(codigo)
    p   = {
        'dataInicial': data_inicial,
        'dataFinal'  : data_final
    }
    response = requests.get(url, params=p)
    if response.status_code == 200:
        dados = pd.DataFrame(response.json())
    else:
        print('Erro na requisição:', response.status_code)
    dados['data'] = pd.to_datetime(dados['data'], format="%d/%m/%Y")
    dados.set_index('data', inplace=True)
    dados['valor'] = pd.to_numeric(dados['valor'])
    dados
    return dados
```

```python
ipca = get_bc(433, data_inicial='01/01/2016')  # Série temporal da IPCA (mês)
cdi  = get_bc(12,  data_inicial='01/01/2016')  # Série temporal da taxa CDI (dia)
```

```python
cdi.sample(6).sort_index() # os valores estão em %
```

Abaixo, temos dois patrimonios iniciais de 100 reais acrescidos de mais 100 reais a cada mês. Um deles rendendo 100% de uma CDI de liquidez diária.

```python
patrimonio = 100
patrimonio2 = 100
serie = [patrimonio]
serie2 = [patrimonio2]
mes = 1

for dia in cdi.index[:-1]:
    if mes != dia.month:
        mes = dia.month
        patrimonio  += 100
        patrimonio2 += 100
    i = cdi[cdi.index == dia]['valor'][0]
    patrimonio = patrimonio*(1+i/100)
    serie.append(patrimonio)
    serie2.append(patrimonio2)
    
patrimonio
```

Aqui plotamos um gráfico para comparar a diferença entre guardar dinheiro e investir em renda-fixa como feito acima. Vemos que a quantia investida parece crescer exponencialmente, o que é esperado em um regime de juros compostos. 

```python
fig1 = plt.figure(figsize=(12,6), dpi=100, facecolor='moccasin')

axes1 = fig1.add_axes([0, 0, 1, 1])
axes1.plot(cdi.index, serie, 'y', label='valor investido')
axes1.plot(cdi.index, serie2, 'k', label='valor depositado')
axes1.set_ylabel('Valor (R$)')
axes1.legend(loc=0)
```

```python
lucro = np.array(serie)-np.array(serie2)

fig1 = plt.figure(figsize=(8,3), dpi=100, facecolor='moccasin')

axes1 = fig1.add_axes([0, 0, 1, 1])
axes1.plot(cdi.index, lucro, color='mediumaquamarine', label='lucro')
axes1.set_ylabel('Valor (R$)')
axes1.set_xlabel('ano')
axes1.legend(loc=0)
```

# Renda Variável

```python
periodo = ('2016-01-01' <= papel.index) #---------> filtro
fig1 = plt.figure(figsize=(12,10), dpi=100, facecolor='moccasin')

plt.plot(papel[periodo]['Close'], 'k', label='fechamento')   
plt.ylabel('Valor (R$)')
plt.legend(loc=0)
```

```python
mes = 0

dividendos = [0]
acoes = []
investido = []

for dia in papel.index:
    acoes_do_mes = 0
    if dia.month != mes:
        mes = dia.month
        acoes_do_mes = 1
        while (acoes_do_mes+1)*papel.loc[dia]['Open'] <= 100: 
            acoes_do_mes += 1
        investido.append(acoes_do_mes*papel.loc[dia]['Open'])
        acoes.append(acoes_do_mes)
    else:
        acoes.append(0)
        investido.append(0)
    DY = 0
    if papel.loc[dia]['Dividends'] > 0.0:
        DY = papel.loc[dia]['Dividends']*(sum(acoes)-acoes_do_mes)
    dividendos.append(DY)
    
dividendos = pd.Series(dividendos)
acoes      = pd.Series(acoes)
investido  = pd.Series(investido)

dividendos.index = papel.index
acoes.index      = papel.index
investido.index  = papel.index
```

```python
fig1 = plt.figure(figsize=(12,6), dpi=100, facecolor='moccasin')

axes1 = fig1.add_axes([0, 0, 1, 1])
axes1.plot(cdi.index, lucro, 'y', label='renda acumulada de renda fixa')
axes1.plot(dividendos.index, dividendos.cumsum(), 'k', label='soma dos dividendos recebidos')
axes1.set_ylabel('Valor (R$)')
axes1.legend(loc=0)
```
