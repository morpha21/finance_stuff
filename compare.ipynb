# Renda Fixa:
Aqui, faremos uma simulação para comparar o investimento de R$ 100,00 ao início de cada mês em uma aplicação de renda fixa que rende diariamente 100% da taxa CDI com simplesmente guardar este dinheiro.   


```python
renda_fixa = 0    # esse patrimônio simulará um investimento em renda fixa
serie = [renda_fixa]

armazenado = 0   # esse patrimônio simulará depósitos simples
serie2 = [armazenado]

mes = 0
investimento = 100 # investimento mensal

for dia in cdi.index[:-1]:
  #--------------------------------# investe ou guarda 100 reais
    if mes != dia.month:           # na virada do mês.
        mes = dia.month            #
        renda_fixa += investimento #
        armazenado += investimento #
  #--------------------------------#
    i = cdi.loc[dia][0]/100 # taxa de juros do dia
    renda_fixa = renda_fixa*(1+i) # capital rende diariamente
    serie.append(renda_fixa)
    serie2.append(armazenado)

serie = pd.Series(serie)
serie2 = pd.Series(serie2)
serie.index, serie2.index = cdi.index, cdi.index
```

# Gráfico 3: 
Comparação dos montantes obtidos pela simulação realizada acima.

```python
ff.dataplot([serie, serie2],
            ['valor investido','valor depositado'])
```

# Gráfico 4: 
Diferença, em cada dia, entre os montantes calculados acima. 

```python
rendimentos = serie-serie2

ff.dataplot([rendimentos], ['rendimentos acumulados'])
```

# Renda Variável
Aqui, a fim de comparar os rendimentos obtidos por renda fixa no método acima com os ganhos por dividendos de uma ação, vamos simular como teria sido a evolução dos ganhos por dividendos de ações compradas mensalmente a partir de 01/01/2016. Na simulação, foram compradas ações todos os meses. No mínimo uma ação, ou quantas fossem possíveis comprar com 100 reais mais os dividendos já obtidos.

Para ficar claro, aqui vamos rodar a seguinte simulação: 
1. Todos os meses, compra-se pelo menos uma ação ou quantas ações forem possíveis com R$ 100,00 mais trocos e dividendos guardados. 
2. Guarda-se o troco, se houver. 
3. No dia em que se receber dividendos, junta-se os dividendos ao troco e, se for o suficiente para comprar ações, compra quantas forem possíveis e guarda-se o troco. 


```python
mes = 0
dinheiro = 0

dividendos = pd.Series(0, index=papel.index)
dividendos_acumulados = pd.Series(0, index=papel.index)
acoes_compradas = pd.Series(0, index=papel.index)
acoes_acumuladas = pd.Series(0, index=papel.index)
investido =  pd.Series(0, index=papel.index)


for dia in papel.index:
#-------------------------------------------------# Início de Mês
    if dia.month != mes:                          #
        mes = dia.month                           #
        acoes=1                                   #
        dinheiro += 100                           #
        if papel['Open'][dia] < dinheiro:         #
            acoes = dinheiro//papel['Open'][dia]  #
            dinheiro -= acoes*papel['Open'][dia]  #
        else:                                     #
            dinheiro = 0                          #
        acoes_compradas[dia] = acoes              #
#-------------------------------------------------# 
    acoes_acumuladas[dia] = acoes_compradas.sum()
    if papel['Dividends'][dia] > 0:
        dinheiro += papel['Dividends'][dia]*acoes_acumuladas[dia]  # usa valores ganhos em dividendos para comprar mais ações
        dividendos[dia] = papel['Dividends'][dia]*acoes_acumuladas[dia]
    dividendos_acumulados[dia] = dividendos.sum()
```

```python
ff.dataplot([dividendos_acumulados, rendimentos], ['dividendos recebidos', 'rendimentos acumulados (renda fixa)'])
```
